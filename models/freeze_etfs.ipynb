```python
# DO NOT TOUCH

import pandas as pd
import yfinance as yf

etf_pairs = [('VO','IWB'), 
         ('VO','SPY'), 
         ('VO','SCHX'), 
         ('VO','SPLG'), 
         ('VO', 'IVV'),
         ('VO', 'VOO'), 
         ('DVY','IWR'), 
         ('SCHV', 'SPHQ'), 
         ('IWR','DIA'), 
         ('DVY','OEF'), 
         ('VTV', 'SPHQ'), 
         ('RSP', 'DIA'), 
         ('VO', 'DIA'), 
         ('DVY', 'FVD'), 
         ('DVY','VYM')] # closing

downloadable_tickers = sorted(set([etf for pair in etf_pairs for etf in pair]))

train_start = "2015-01-01"
train_end = "2020-01-01"
test_start = "2020-01-02"
test_end = "2024-12-31"

train_prices = yf.download(downloadable_tickers, start = train_start, end = train_end)["Close"].dropna()
test_prices = yf.download(downloadable_tickers, start = test_start, end = test_end)["Close"].dropna()

train_prices.to_csv("etf_train_prices.csv")
test_prices.to_csv("etf_test_prices.csv")
```
