```python
import yfinance as yf
import pandas as pd

tickers = ['MO', 'F', 'STX', 'BTI', 'HD', 'LOW', 'HWM', 'BK', 'CMI', 'NVDA', 'AAPL', 'MSFT', 'AVGO',
           'JPM', 'XOM', 'GS', 'CAT', 'WBD', 'ALB', 'WDC', 'JNJ', 'WMT', 'ABBV', 'GE', 'MA', 'CRH', 'NEM',
           'AXP', 'V', 'GOOG', 'META', 'TSLA', 'BRK-B', 'LLY', 'AMZN', 'NFLX', 'GOOGL', 'BE', 'FN', 'KTOS',
           'ROKU', 'CRSP']

train_start = "2015-01-01"
train_end = "2020-01-01"
test_start = "2020-01-02"
test_end = "2024-12-31"

train_prices = yf.download(tickers, start = train_start, end = train_end)["Close"].dropna()
test_prices = yf.download(tickers, start = test_start, end = test_end)["Close"].dropna()

train_prices.to_csv("stock_train_prices.csv")
test_prices.to_csv("stock_test_prices.csv")
```
