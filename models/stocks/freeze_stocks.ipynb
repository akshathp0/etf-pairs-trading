```python
import yfinance as yf
import pandas as pd

tickers = ['MO', 'F', 'STX', 'BTI', 'HD', 'LOW', 'BK', 'CMI', 'NVDA', 'AAPL', 'MSFT', 'AVGO',
           'JPM', 'XOM', 'GS', 'CAT', 'WBD', 'ALB', 'WDC', 'JNJ', 'WMT', 'ABBV', 'GE', 'MA', 'CRH', 'NEM',
           'AXP', 'V', 'GOOG', 'META', 'TSLA', 'BRK-B', 'LLY', 'AMZN', 'NFLX', 'GOOGL', 'FN', 'KTOS',
           'PWR', 'TDG']

train_start = "2015-01-01"
train_end = "2020-01-01"
test_start = "2020-01-02"
test_end = "2024-12-31"

train_prices = yf.download(tickers, start = train_start, end = train_end)["Close"]
test_prices = yf.download(tickers, start = test_start, end = test_end)["Close"]

train_start_idx = train_prices.index.min()

bad_tickers = [
    p for p in train_prices.columns
    if train_prices[p].first_valid_index() is None
    or train_prices[p].first_valid_index() > train_start_idx
]

print("Remove:")
print(bad_tickers)

train_prices = train_prices.dropna()
test_prices = test_prices.dropna()

train_prices.to_csv("stock_train_prices.csv")
test_prices.to_csv("stock_test_prices.csv")
```

```python
import pandas as pd
import yfinance as yf

new_tickers = ['HOUS', 'COMM', 'VSAT']

train = pd.read_csv('stock_train_prices.csv', index_col = 0, parse_dates = True)
test = pd.read_csv('stock_test_prices.csv', index_col = 0, parse_dates = True)

train_start = "2015-01-01"
train_end = "2020-01-01"
test_start = "2020-01-02"
test_end = "2024-12-31"

new_train = yf.download(new_tickers, start = train_start, end = train_end)['Close']
new_test = yf.download(new_tickers, start = test_start, end = test_end)['Close']

train_start_idx = train.index.min()

bad_new_tickers = [
    p for p in new_train.columns
    if new_train[p].first_valid_index() is None
    or new_train[p].first_valid_index() > train_start_idx
]

print("Remove:")
print(bad_new_tickers)
```

```python
train = pd.concat([train, new_train], axis = 1).dropna()
test = pd.concat([test, new_test], axis = 1).dropna()

train.to_csv('stock_train_prices.csv')
test.to_csv('stock_test_prices.csv')
```
